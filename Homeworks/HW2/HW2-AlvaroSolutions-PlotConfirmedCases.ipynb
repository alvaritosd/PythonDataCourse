```python

# resize this figure here

plt.figure(figsize=(5,5))
plt.plot(data_italy.index,data_italy.values,'.-')
plt.xticks(rotation=90) # rotate the xticks so text is not so tight
plt.xlabel('calendar date')
plt.show()
```

```python
import numpy as np

df_spain = df[df['Country/Region'] == 'Spain']
df_spain = df_spain.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_spain = df_spain.drop(columns=['Lat','Long','Province/State'])

data_spain = df_spain.loc['Spain','2/20/20':'4/13/20']

df_colombia = df[df['Country/Region'] == 'Colombia']
df_colombia = df_colombia.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_colombia = df_colombia.drop(columns=['Lat','Long','Province/State'])

data_colombia = df_colombia.loc['Colombia','2/20/20':'4/13/20']

df_chile = df[df['Country/Region'] == 'Chile']
df_chile = df_chile.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_chile = df_chile.drop(columns=['Lat','Long','Province/State'])

data_chile = df_chile.loc['Chile','2/20/20':'4/13/20']

df_france = df[df['Country/Region'] == 'France']
df_france = df_france.set_index('Country/Region', drop = True) # set pandas dataframe index to country
df_france = df_france.drop(columns=['Lat','Long','Province/State'])

data_france = df_france.loc['France','2/20/20':'4/13/20']


data_world=[data_italy, data_us, data_spain, data_colombia, data_chile, data_france]

plt.figure(figsize=(8,4))
plt.plot(data_italy, label='Italy')
plt.plot(data_us, label='US')
plt.plot(data_spain, label='Spain')
plt.plot(data_colombia, label='Colombia')
plt.plot(data_chile, label='Chile')
#plt.plot(data_france, label='France')


for data in data_world:
    plt.yscale('log')
    plt.legend()
    plt.ylim(1,700000)
    plt.xticks(np.arange(0,60,5), rotation=45)
    plt.xlabel('date',fontsize=14)
    plt.ylabel('number of confirmed cases',fontsize=14)    

plt.legend(fontsize=12)
plt.savefig('alvaroplot.jpg')

# save your figure
# fig.savefig('myplot.jpg')
```
